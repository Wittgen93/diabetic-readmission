Ячейка 1. Импорт и настройки

```python
import pandas as pd
import numpy as np

# Настройки отображения
pd.set_option('display.max_columns', 100)
pd.set_option('display.width', 120)
```

Ячейка 2. Загрузка данных и первичный просмотр

```python
# Загружаем CSV
df = pd.read_csv("../data/raw/diabetic_data_DSMED.csv")

# Посмотрим первые строки
display(df.head())

# Информация о столбцах и типах
df.info()
```

 Ячейка 3. Описательная статистика по числовым признакам

```python
# Базовая статистика
stats = df.describe().T
display(stats)
```

Ячейка 4. Поиск и замена “?”

```python
# Считаем, сколько “?” в каждом столбце
q_counts = {col: (df[col] == '?').sum() for col in df.columns}
q_counts = {k:v for k,v in q_counts.items() if v>0}
print("Столбцы с '?' и их количество:\n", q_counts)

# Заменяем все '?' на NaN
df.replace('?', np.nan, inplace=True)
```

Ячейка 5. Оценка доли пропусков и удаление столбцов с >80 % NaN

```python
# Процент пропусков
missing_pct = df.isna().mean().mul(100).round(2)
display(missing_pct[missing_pct>0])

# Столбцы с >80% пропусков
to_drop = missing_pct[missing_pct > 80].index.tolist()
print("Удаляем столбцы с >80% пропусков:", to_drop)

# Удаляем
df.drop(columns=to_drop, inplace=True)
```

Ячека 6. Заполнение оставшихся пропусков

```python
# Список категориальных, где пропуски заполняем 'Unknown'
fill_cols = ['race','payer_code','medical_specialty','diag_1','diag_2','diag_3']
for col in fill_cols:
    df[col] = df[col].fillna('Unknown')

# Проверим, что пропусков больше нет
print("Осталось NaN:", df.isna().sum().sum())
```

Ячейка 7. Преобразование age в числовой формат

```python
# Функция для перевода "[70 - 80)"  - > 75
def age_to_mean(age_str):
    lo, hi = age_str.strip('[]()').split(' - ')
    return (int(lo) + int(hi)) / 2

df['age_num'] = df['age'].apply(age_to_mean)

# Смотрим распределение нового признака
display(df['age_num'].describe())
```

Ячейка 8. Удаление служебных столбцов

```python
# Эти столбцы не будут использоваться в моделях
drop_meta = ['encounter_id', 'patient_nbr', 'age']
df.drop(columns=drop_meta, inplace=True)

# Финальная проверка структуры
df.info()
```

Ячейка 9. Сохраняем очищенный DF

```python
import os
os.makedirs("../data/processed", exist_ok=True)
df.to_csv("../data/processed/df_clean.csv", index=False)
```
